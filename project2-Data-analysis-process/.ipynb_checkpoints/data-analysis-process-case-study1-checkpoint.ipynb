## Introduction
In this dataset, we will analyze breast cancer winsconsin dataset, we will analyze the dataset and predict whether the cancer is benign or malignant

```python
import pandas as pd
%matplotlib inline
```

```python
data = pd.read_csv('data.csv')
```

```python
data.describe()
```

```python
data.nunique()
```

```python
data.describe(include=['O'])
```

```python
columns = data.columns
```

```python
print(columns)
```

```python
# cleaning the dataset columns
cleaned_columns = []
def clean_columns(column_name):
    column_name = column_name.replace(' ','_')
    return column_name
```

```python
for item in columns:
    cleaned_column = clean_columns(item)
    cleaned_columns.append(cleaned_column)
```

```python
data.columns = cleaned_columns
```

```python
data.head()
```
