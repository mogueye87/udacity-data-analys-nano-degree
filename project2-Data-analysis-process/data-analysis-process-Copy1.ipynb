## Data Cleaning process
* Clean the columns replacing whitespaces and underscore
* Checking for missing values
* In correct datatypes and format
* Checking for duplicates rows and removes them if any

```python
red_wine_columns = red_wine.columns
```

```python
white_wine_columns = white_wine.columns
```

```python
print(red_wine_columns)
```

```python
print(white_wine_columns)
```

```python
# cleaning the dataset columns
cleaned_columns = []
def clean_columns(columns):
    """
     This function takes as an argument an array of strins
     and check whether the string contains whitespaces
     and replace any by underscore.
    """
    for item in columns:
        if item.find(' ') != -1:
            column_name = item.replace(' ','_')
            cleaned_columns.append(column_name)
        else:
            cleaned_columns.append(item)
    return cleaned_columns
```

```python
# cleaning data sets columns
results = clean_columns(red_wine.columns)
print(results)
```

```python
red_wine.columns = results
```
